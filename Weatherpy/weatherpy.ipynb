```python
# Dependencies
import csv
import numpy as np
import time
import matplotlib.pyplot as plt
import requests
import pandas as pd
from api_keys import api_key
from citipy import citipy
# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

#Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

#Print the city count to confirm sufficient count
print(len(cities))
```

```python
print(len(city_exist))
print(len(lat))
print(len(temp))
print(len(humidity))
print(len(wind))
print(len(cloudiness))
```

```python
# create a data frame from cities, lat, and temp
weather_dict = {
    "city": city_exist,
    "lat": lat,
    "temp": temp,
    "humidity":humidity,
    "wind speed":wind,
    "cloudiness":cloudiness
}
weather_data = pd.DataFrame(weather_dict)
weather_data.head()
```

```python
# Build a scatter plot for each data type
plt.scatter(weather_data["lat"], weather_data["temp"], marker="o", color='red', alpha=0.5)
# Incorporate the other graph properties
plt.title("Temperature in Cities by Latitude")
plt.ylabel("Temperature (Celsius)")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("TemperaturebyLatitude.png")

# Show plot
plt.show()
```

```python
plt.scatter(weather_data["lat"], weather_data["humidity"], marker="s", color='blue', alpha=0.5)
# Incorporate the other graph properties
plt.title("Humidity in Cities by Latitude")
plt.ylabel("Humidity")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("HumiditybyLatitude.png")

# Show plot
plt.show()
```

```python
plt.scatter(weather_data["lat"], weather_data["wind speed"], marker="d", color='green', alpha=0.5)
# Incorporate the other graph properties
plt.title("Wind speed in Cities by Latitude")
plt.ylabel("Wind Speed")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("windspeedbyLatitude.png")

# Show plot
plt.show()
```

```python
plt.scatter(weather_data["lat"], weather_data["cloudiness"], marker="+", color='black', alpha=0.5)
# Incorporate the other graph properties
plt.title("Cloudiness in Cities by Latitude")
plt.ylabel("Cloudiness")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("cloudinessbyLatitude.png")

# Show plot
plt.show()
```
